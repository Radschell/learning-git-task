```python
import numpy as np
from scipy import stats

# Wygeneruj 100 punktów danych o średniej równej 3 i odchyleniu standardowym równej 1
data_100 = np.random.normal(3, 1, 100)

# Oblicz średnią
average = np.mean(data_100)

# Oblicz medianę
median = np.median(data_100)

# Oblicz wartość modalną
mode = stats.mode(data_100)[0][0]

# Oblicz kwantyle
q1, q2, q3 = np.percentile(data_100, [25, 50, 75])

# Oblicz zakres
range_ = np.max(data_100) - np.min(data_100)

# Oblicz odchylenie międzykwartylowe (IQR)
iqr = q3 - q1

# Oblicz wariancję
variance = np.var(data_100)

# Oblicz odchylenie standardowe
std_dev = np.std(data_100)

# Wyświetl wyniki
print("Średnia:", average)
print("Mediana:", median)
print("Wartość modalna:", mode)
print("Q1:", q1)
print("Q2:", q2)
print("Q3:", q3)
print("Zakres:", range_)
print("Odchylenie międzykwartylowe (IQR):", iqr)
print("Wariancja:", variance)
print("Odchylenie standardowe:", std_dev)

import numpy as np
from scipy import stats

# Wygeneruj 10 000 punktów danych o średniej równej 3 i odchyleniu standardowym równej 1
data_10000 = np.random.normal(3, 1, 10000)

# Oblicz średnią
average = np.mean(data_10000)

# Oblicz medianę
median = np.median(data_10000)

# Oblicz wartość modalną
mode = stats.mode(data_10000)[0][0]

# Oblicz kwantyle
q1, q2, q3 = np.percentile(data_10000, [25, 50, 75])

# Oblicz zakres
range_ = np.max(data_10000) - np.min(data_10000)

# Oblicz odchylenie międzykwartylowe (IQR)
iqr = q3 - q1

# Oblicz wariancję
variance = np.var(data_10000)

# Oblicz odchylenie standardowe
std_dev = np.std(data_10000)

# Wyświetl wyniki
print("Średnia:", average)
print("Mediana:", median)
print("Wartość modalna:", mode)
print("Q1:", q1)
print("Q2:", q2)
print("Q3:", q3)
print("Zakres:", range_)
print("Odchylenie międzykwartylowe (IQR):", iqr)
print("Wariancja:", variance)
print("Odchylenie standardowe:", std_dev)
```
